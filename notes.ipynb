```python
class Student:
    school_name = "Python Academy"
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    @classmethod
    def from_birth_year(cls, name, birth_year):
        """Create a Student instance using birth year instead of age"""
        age = 2024 - birth_year
        return cls(name, age)
    
    @classmethod
    def change_school(cls, new_school):
        """Change school name for all instances"""
        cls.school_name = new_school

# Example usage
student1 = Student("Alice", 20)
print(f"{student1.name} is {student1.age} years old, studies at {student1.school_name}")

# Using classmethod to create instance
student2 = Student.from_birth_year("Bob", 2000)
print(f"{student2.name} is {student2.age} years old, studies at {student2.school_name}")

# Using classmethod to modify class variable
Student.change_school("Advanced Python School")
print(f"{student1.name} now studies at {student1.school_name}")
```
